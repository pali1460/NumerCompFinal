```python
def lagrange(x, y):
    @np.vectorize
    def p(t):
        from numpy import prod
        m = len(x) - 1
        w = 0
        for i in range(m):
            w += y[i] * (prod(t - x[:i]) * prod(t - x[i+1:])
                         / (prod(x[i] - x[:i]) * prod(x[i] - x[i+1:])))
        w += y[m] * prod(t - x[:m]) / prod(x[m] - x[:m])
        return w
    return p

x = np.linspace(1893,2023,13)
y = M1_Mean10
p = lagrange(x, y)
xx = np.linspace(1890,2020)

ax = plt.gca()
#ax.set_xlim([xmin, xmax])
ax.set_ylim([0, 3])

plt.plot(x, y, 'o', label='January Data Points')
plt.plot(xx, p(xx), label='Lagrange Interpolation p(x)')

#Lagrange isn't very useful. 


plt.title("January Data and Interpolation")
plt.xlabel("Year")
plt.ylabel("10-Year Average Precipitation (Inches)")
plt.legend(loc='upper left');
```

```python
#Final chosen method: Cubic Spline Interpolation

#By taking averages, we can get better interpolation. We will take averages for every 10 years. 

#10 years both has enough points and not too many points, so we will use this for further interpolation. 

#Auto-generate the years, using the median point of each range

Years = []

for i in range(13):
    Years.append(1893 + 5 + 10*i)

def yearMean(data):
    yearMean = []
    for i in range(13):
        yearMean.append(np.mean(data[0 + 10*i: 10 + 10*i]))
    return yearMean


def realAvg(data, year):
    return np.mean(data[year - 5 - 1893: year + 4 - 1893])
```

```python
useMonth = M1

useData = yearMean(useMonth)

plt.plot(Years, useData, 'o', label = 'Data Points')

x_interp = np.linspace(np.min(Years), np.max(Years), 50)
y_cubicbc = CubicSpline(Years, useData, bc_type = "natural")

#Plot it
plt.plot(x_interp, y_cubicbc(x_interp), 'blue', label = 'Cubic Spline BC')

#Extrapolate, calculate error. 
givenYear = 2019

print("Projected rainfall from the year " 
      + str(givenYear - 5) + " to "
      + str(givenYear + 4) + " is " 
      + str(np.round(y_cubicbc(givenYear), 3)))

print("Real rainfall from the year " 
  + str(givenYear - 5) + " to "
  + str(givenYear + 4) + " is " 
  + str(np.round(realAvg(useMonth, givenYear), 3)))

print("Absolute error is " + 
     str(np.round(abs(y_cubicbc(givenYear) - realAvg(useMonth, givenYear)), 5))
     + ", relative error is " + 
     str(np.round(abs(y_cubicbc(givenYear) - realAvg(useMonth, givenYear)) / realAvg(useMonth, givenYear), 5))
     )

#Label graph
plt.title("January Data and Interpolation")
plt.xlabel("Year")
plt.ylabel("10-Year Average Precipitation (Inches)")

plt.legend()
```
